```python
#instead of dropping converting the text info into boolean 
temp= pd.get_dummies(data, columns=['Genre-Defind'])
```

```python
temp_1 = pd.get_dummies(temp,columns = ['Release_Date(N/LW/Festive)','Movie_Content','Dir_CAT','LEAD_ACTOR_CAT','Production_House_CAT','Music_Dir_CAT'])
```

```python
modified_data.drop('Movie_Name',axis = 1,inplace = True)
```

```python
#Splitting the data
from sklearn.model_selection import train_test_split
X = modified_data.drop('Success/Faliure',axis = 1)
y = modified_data['Success/Faliure']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 3) 

```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(criterion = 'entropy',random_state = 15)
dt_model.fit(X_train,y_train)
```

```python
dt_model.score(X_train,y_train)
```

```python
dt_model.score(X_test,y_test)
```

```python
# Regularised_desion_tree
rdt_model = DecisionTreeClassifier(criterion = 'entropy',max_depth = 1, min_samples_leaf = 2 ,random_state = 15)
rdt_model.fit(X_train,y_train)
```

```python
rdt_model.score(X_train,y_train)
```

```python
rdt_model.score(X_test,y_test)
```


# RAndom Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfcl = RandomForestClassifier(n_estimators = 10,criterion = 'entropy')
rfcl.fit(X_train,y_train)
```

```python
rfcl.score(X_train,y_train)
```

```python
rfcl.score(X_test,y_test)
```

```python
rfcl.predict(X_test)
```

# Bagging

```python
from sklearn.ensemble import BaggingClassifier
bgcl = BaggingClassifier(base_estimator = dt_model,n_estimators=10,max_samples = 10,bootstrap = True,oob_score = True)
```

```python
bgcl.fit(X_train,y_train)
```

```python
bgcl.score(X_train,y_train)
```

```python
bgcl.oob_score_
```

```python
bgcl.score(X_test,y_test)
```

# AdaBoosting

```python
from sklearn.ensemble import AdaBoostClassifier
abcl = AdaBoostClassifier(base_estimator = dt_model,n_estimators = 10,random_state = 10)
abcl.fit(X_train,y_train)
```

```python
abcl.score(X_train,y_train)
```

```python
abcl.score(X_test,y_test)
```

# Gradient Boosting 

```python
from sklearn.ensemble import GradientBoostingClassifier
gbcl = GradientBoostingClassifier(n_estimators = 20)
gbcl.fit(X_train,y_train)
```

```python
gbcl.score(X_train,y_train)
```

```python
gbcl.score(X_test,y_test)
```
